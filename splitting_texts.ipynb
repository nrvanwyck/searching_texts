```python
from spacy.lang.en import English

filename = 'oscar_wilde_intentions.txt'

with open(filename) as file:
    raw_text = file.read()
    
raw_text = raw_text.replace('CYRIL.', 'CYRIL:').replace(
    'VIVIAN.', 'VIVIAN:').replace('GILBERT.', 'GILBERT:').replace(
    'ERNEST.', 'ERNEST:')
    
nlp = English()
nlp.add_pipe(nlp.create_pipe('sentencizer'))
doc = nlp(raw_text)
sentences = [sent.string.replace('\n', ' ').strip() for sent in doc.sents]
```

```python
for sentence in sentences[8:2559]:
    print(sentence)
    print()
    print()
```
